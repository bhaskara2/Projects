```python
#Visualizations on IRIS dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#seaborn is a compliment/extension to matplotlib, it works well on pandas data frames and arrays than Matplotlib

new_names=['ID','Sepal Length','Sepal Width','Petal Length','Petal Width','Species']
iris=pd.read_csv('C:/Users/Bhaskara/Downloads/Summer/Work/Python/IRIS.csv',header=0,names=new_names)
#iris_new = iris[[specify column names]].copy()
iris.head()


```

```python
#scatter plot 
#iris.plot(x="Sepal Length",y="Sepal Width",kind="scatter",)
#plt.show()

#colour the points by the species

sns.FacetGrid(iris,hue='Species',size=5) \
.map(plt.scatter, "Sepal Length", "Sepal Width") \
   .add_legend()

#we can see that Setosa is feauturs are well seperated frm the other two species.

#Explore Different Seaborn visual techniques
sns.boxplot(x='Species',y='Sepal Length',data=iris)
plt.show()
```

```python
sns.boxplot(x='Species',y='Sepal Width',data=iris)
plt.show()
#the sepal width of setosa is higher than other species ans sepal length is low compared to other other two species.
```

```python
#by using pairplot we can see how the sepal and petal dimensions are varying across the different species infer the difference 
#between species

sns.pairplot(iris.drop('ID',axis=1),hue='Species',size=3)
plt.show()
```

```python
#change the diagnol plot (Histograms) to KDE(Kernel Density Plots)
sns.pairplot(iris.drop('ID',axis=1),hue='Species',size=3,diag_kind='kde')
plt.show()

#The petal lengths for the species are quite distict and well seperated.
```

```python
plt.figure(figsize=(7,4)) 
iris_new=iris.drop('ID',axis=1)
sns.heatmap(iris_new.corr(),annot=True,cmap='cubehelix_r') #draws  heatmap with input as the correlation matrix calculted by(iris.corr())
plt.show()

#there seems to be high correlation between petal lengtn and petal width and also there is correlation between different
#parameters of sepals aswell.
```
